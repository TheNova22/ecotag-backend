```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("completePortList copy.csv")
```

```python
df.head()
```

```python
df = df.drop(["Ch", "IATA", "Remarks"], axis = 1)
```

```python
df.head()
```

```python
df = df[df["Coordinates"].notna()]
```

```python
df.head()
```

```python
import utm
cds = np.array(df["Coordinates"])

lats = []

longs = []

for i in range(len(cds)):
    arr = cds[i].split()

    lats.append(float(arr[0][:-3] + "." + arr[0][-3:-1]))

    longs.append(float(arr[1][:-3] + "." + arr[1][-3:-1]))


df["Latitude"] = lats

df["Longitude"] = longs

df.head()
```

```python
water = df[df["Function"].str.contains(pat = "1")]

water = water.reset_index()

water.head()
```

```python
rail = df[df["Function"].str.contains(pat = "2")]

rail = rail.reset_index()

road = df[df["Function"].str.contains(pat = "3")]

road = road.reset_index()

air = df[df["Function"].str.contains(pat = "4")]

air = air.reset_index()
```

```python
air.head()
```

```python
water.to_csv("ports.csv")

road.to_csv("road_terminals.csv")

rail.to_csv("rail_terminals.csv")

air.to_csv("airports.csv")
```
